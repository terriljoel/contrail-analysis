```python
import os
from pathlib import Path
import cdsapi

def find_project_root(start: Path) -> Path:
    p = start.resolve()
    for parent in [p] + list(p.parents):
        if (parent / "data").exists():
            return parent
    return p

ROOT = find_project_root(Path.cwd())

ROOT
```

```python




# Day you processed flights for
YEAR  = "2025"
MONTH = "01"
DAY   = "13"

# Output folder
area = "states_europe"
OUT_DIR = ROOT / f"data/raw/weather_ERA5/{area}" / f"{YEAR}-{MONTH}-{DAY}"
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Europe bbox: [N, W, S, E]

AREA = [72, -15, 35, 35]

# Cruise-ish pressure levels (hPa)
PLEVS = ["350", "300", "250", "225", "200"]

# Hourly times
TIMES = [f"{h:02d}:00" for h in range(24)]

target = OUT_DIR / f"era5_pl_europe_T_q.nc"

c = cdsapi.Client()

dataset = "reanalysis-era5-pressure-levels"
request = {
    "product_type": ["reanalysis"],
    "variable": ["temperature", "specific_humidity"],
    "year": [YEAR],
    "month": [MONTH],
    "day": [DAY],
    "time": TIMES,
    "pressure_level": PLEVS,
    "area": AREA,
    "data_format": "netcdf",
    "download_format": "unarchived",
}

c.retrieve(dataset, request, str(target))
print("Saved:", target)
```
